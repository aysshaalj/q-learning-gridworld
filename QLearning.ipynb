```python
import numpy as np
import random
import matplotlib.pyplot as plt
from matplotlib import colors
from IPython.display import clear_output
import time

print("Q-LEARNING TUTORIAL")
print("="*50)

#grid world setup
print("\n SETTING UP THE ENVIRONMENT")
print("  -----------------------------")

grid_size = 5
goal_state = (4, 4)
start_state = (0, 0)
actions = ['up', 'down', 'left', 'right']

#define obstacles
obstacles = [(1,1), (2,2), (3,1), (1,3)]

print(f"grid_size: {grid_size}*{grid_size}")
print(f"Start: {start_state}, Goal: {goal_state}")
print(f"Obstacles: {obstacles}")
print(f"Possible Actions: {actions}")

#initialize Q-table
Q = np.zeros((grid_size, grid_size, len(actions)))
print(f"Q-table shape: {Q.shape} <- (rows, columns, actions)")
print("This meas: 5 rows x 5 columns x 4 possible actions")

#hyperparameters
print("\n Setting learning parameters")
print("  -----------------------------")

alpha = 0.1 #learning rate: how much we learn from new experiences
gamma = 0.9 #Discount factor: how much we calue future rewards
epsilon = 0.3 #Eploration rate: chance of trying random actions
episodes = 500 # number of training games

print(f"Learning Rate: {alpha} - how fast we learn")
print(f"Discount Factor: {gamma} - how much we care about future rewards")
print(f"exploration rate: {epsilon} - chance of exploring vs using what we know")
print(f"traing episodes: {episodes} -  how many times we'll practice")

# helper functions with explanations
print("\n how the agent makes decisions")
print("  -----------------------------")

def choose_action (state):
  """
  Chooses an action based on epsilon-greedy strategy.
  Either explores (random action) or exploits (best known action).
  """
  if random. uniform(0, 1) < epsilon:
    #EXPLORATION: Try a random action
    action = random. choice (range(len(actions)))
    print(f" & Exploringl Random action: {actions[action]}")
    return action
  else:
    #EXPLOITATION: Use the best action we know
    action = np.argmax(Q[state[0], state[1]])
    print(f"Exploiting! Best known action: {actions[action]}")
    return action

def step(state, action):
  """
  Take action and see what happens in the environment
  """
  i, j = state

  #move based on states
  if actions[action] == 'up':
    i = max(i-1, 0)
  elif actions[action] == 'down':
    i = min(i+1, grid_size-1)
  elif actions[action] == 'left':
    j = max(j-1, 0)
  elif actions[action] == 'right':
    j = min(j+1, grid_size-1)

  #check for obstacles - if we hit one, stay in place
  if(i,j) in obstacles:
    print(f" Hit obstacle! Stay in place at {state}")
    i,j = state
    reward = -5 #big penalty for hitting obstacel
    done = False
  #check iof reached the gial
  elif(i,j) == goal_state:
    reward = 20 #big reward for success
    done = True
    print(f"Reached Goal!! Got reward = {reward}")
  else:
    reward = -1 #small penalty for move (encourage efficiency)
    done = False

  return (i,j), reward, done

def update_q_table(state, action, reward, next_state):
  current_q = Q[state[0], state[1], action]
  max_next_q = np.max(Q[next_state[0], next_state[1]])

  #Q-learninf update rule
  new_q = current_q + alpha * (reward + gamma * max_next_q - current_q)
  Q[state[0], state[1], action] = new_q

  print(f"Learning update:")
  print(f"State {state}, Action: {actions[action]}")
  print(f"Old Q-value: {current_q:.2f} -> New Q-value: {new_q:.2f}")
  print(f"Reward: {reward}, Max Future Q: {max_next_q:.2f}")

#visualization functions
print("\n Creating visualization")
print("  -----------------------------")

def plot_grid(agent_pos, episode=None, step_count=None):
  grid = np.zeros((grid_size, grid_size))

  #mark special cells
  for obs in obstacles:
    grid[obs] = -1 #obstacles
  grid[goal_state] = 2 #goal
  grid[agent_pos] = 1 #agent
  grid[start_state] = 0.5 #starts (diff color)

  #create colromap
  cmap = colors.ListedColormap(['white', 'blue', 'green', 'red', 'yellow'])
  bounds = [-1.5, -0.5, 0.5, 1.5, 2.5, 3.5]
  norm = colors.BoundaryNorm(bounds, cmap.N)

  plt.figure(figsize=(6,6))
  plt.imshow(grid, cmap=cmap, norm=norm)

  #add grid lines and lables
  plt.xticks(range(grid_size))
  plt.yticks(range(grid_size))
  plt.grid(which='both', color='black', linestyle='-', linewidth=2)

  #add text annotations
  for i in range(grid_size):
    for j in range(grid_size):
      if (i, j) == agent_pos:
        plt.text(j, i, 'A', ha='center', va='center', fontsize=20)
      elif (i, j) == goal_state:
        plt.text(j, i, 'G', ha='center', va='center', fontsize=20)
      elif (i, j) in obstacles:
        plt.text(j, i, 'X', ha='center', va='center', fontsize=15)
      elif (i,j) == start_state:
        plt.text(j, i, 'S', ha='center', va='center', fontsize=15)

  title = "Q-Learning Grid World"
  if episode is not None:
    title += f" - Episode {episode}"
  if step_count is not None:
    title += f" - Step {step_count}"
  plt.title(title, fontsize=14, pad=20)
  plt.show()

def show_q_table_preview():
  print("\n Q-TABLE PREVIEW (first 3 states)")
  print("State  |  Up  |  Down  |  Left |  Right ")
  print("-"*45)
  for i in range(min(3, grid_size)):
    for j in range(min(3, grid_size)): # Added inner loop for columns
      if (i, j) not in obstacles and (i, j) != goal_state:
        q_vals = [f"{Q[i, j , a]:6.2f}" for a in range(len(actions))]
        print(f"({i}, {j})  |  {q_vals[0]}  |  {q_vals[1]}  |  {q_vals[2]}  |  {q_vals[3]}  ")


#training the agent
print("\n TRAINING THE AGENT")
print("  -----------------------------")
print("the agent will now learn through trial and error!")
print("watching first 2 episodes in detail")

steps_per_episode = []
successful_episodes = 0

for episode in range(episodes):
  state = start_state
  done = False
  step_count = 0

  #show detailed view of first 2 episodes
  show_details = episode < 2

  if show_details:
    print(f"Episode {episode + 1} - watch learninf happen!")
    print("="*50)
    plot_grid(state, episode = episode+1, step_count = 0)

  while not done and step_count < 50:
    if show_details:
      print(f"\nStep {step_count+1}:")

    #choose and take action
    action = choose_action(state)
    next_state, reward, done = step(state, action)

    #update Qtavle
    update_q_table(state, action, reward, next_state)

    state = next_state
    step_count += 1

    if show_details:
      plot_grid(state, episode = episode+1, step_count=step_count)
      time.sleep(1)

    if done and reward > 0:
      successful_episodes +=1
```
